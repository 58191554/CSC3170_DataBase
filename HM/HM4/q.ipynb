```python
x = [i for i in range(1,20)]
y = []
for i in range(19):
    a = (1+i+1)/2
    print(a, end = " ")
    y.append(a)
z = []
for j in range(19):
    div = 0
    
    for i in range(j+1):
        div += 1/(i+1)
    b = j+1/div
    z.append(b)
    print(b, end=" ")
import matplotlib.pyplot as plt
plt.plot(x,y)
plt.plot(x,z)
```
